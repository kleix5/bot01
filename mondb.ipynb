```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv', index_col='Регион')
df.head()
```

```python
df = df.sort_index()
```

```python
indar = np.unique(np.array(df.index))
```

Нам необходимо удалить ненужные показатели. Для этого мы транспонируем матрицу
подсвечиваем NaN и добавляем столбец, содержащий результат применения логического И
к каждому ряду.

```python
dft = df.T
```

```python
dft['andres'] = dft.apply(lambda row: all(row.isna()), axis=1)
```

```python
df = dft[dft.andres != True].T
```
